```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
y = np.loadtxt('inv_y.txt')
y_hat = np.loadtxt('inv_yhat.txt')
verygood_lk = pd.read_csv('verygood_lk.csv', header = None).values
prettygood_lk = pd.read_csv('prettygood_lk.csv', header = None).values
normal_lk = pd.read_csv('normal_lk.csv', header = None).values
extremelybad_lk = pd.read_csv('extremelybad_lk.csv', header = None).values
```

### 测试样本一（very good）

#### 先分别测试第一个预测点和最后一个预测点对预测值的跟踪情况
#### 然后取早晨（6：00）、中午（12:00）、下午（16:00）的预测的未来四小时功率值进行对比
#### 最后计算RMSE

```python
t = np.arange(len(extremelybad_lk.T))

plt.plot(t, y[96*10+20:96*10+81,0], label = 'TRUE_VALUE')
plt.plot(t, y_hat[96*10+20:96*10+81,0], label = 'ANN_FORECAST')
plt.plot(t, verygood_lk[0,:], label = 'NCEPU_FORECAST')

plt.legend()
```

```python
plt.plot(t, y[96*10+20:96*10+81,15], label = 'TRUE_VALUE')
plt.plot(t, y_hat[96*10+20:96*10+81,15], label = 'ANN_FORECAST')
plt.plot(t, verygood_lk[15,:], label = 'NCEPU_FORECAST')

plt.legend()
```

```python
t1 = np.arange(16)

plt.plot(t1, y[96*10+24,:], label = 'TRUE_VALUE')
plt.plot(t1, y_hat[96*10+24,:], label = 'ANN_FORECAST')
plt.plot(t1, verygood_lk[:,5], label = 'NCEPU_FORECAST')

plt.legend()
```

```python
plt.plot(t1, y[96*10+48,:], label = 'TRUE_VALUE')
plt.plot(t1, y_hat[96*10+48,:], label = 'ANN_FORECAST')
plt.plot(t1, verygood_lk[:,29], label = 'NCEPU_FORECAST')

plt.legend()
```

```python
plt.plot(t1, y[96*10+64,:], label = 'TRUE_VALUE')
plt.plot(t1, y_hat[96*10+64,:], label = 'ANN_FORECAST')
plt.plot(t1, verygood_lk[:,45], label = 'NCEPU_FORECAST')

plt.legend()
```

```python
verygood_test = y[96*10+20:96*10+81]
verygood_ann = y_hat[96*10+20:96*10+81]
verygood_ncepu = verygood_lk.T
verygood_rmse1 = np.sqrt(np.average((verygood_test-verygood_ann)**2,axis = 0))
verygood_rmse2 = np.sqrt(np.average((verygood_test-verygood_ncepu)**2,axis = 0))

verygood_rmse = pd.DataFrame({'ann':verygood_rmse1, 'ncepu':verygood_rmse2})
verygood_rmse
```
